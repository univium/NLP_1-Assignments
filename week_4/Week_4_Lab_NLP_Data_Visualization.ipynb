<b>Stopwords</b> are the English words which does not add much meaning to a sentence. They can safely be ignored without sacrificing the meaning of the sentence. For example, the words like the, he, have etc.(https://www.tutorialspoint.com/python_text_processing/python_remove_stopwords.htm#:~:text=Stopwords%20are%20the%20English%20words,this%20in%20corpus%20named%20corpus.)


We have given you a function to plot stopwords below.

```python
from nltk.corpus import stopwords
import matplotlib.pyplot as plt

def plot_top_stopwords_barchart(text):
    stop=set(stopwords.words('english'))
    
    new= text.str.split()
    new=new.values.tolist()
    corpus=[word for i in new for word in i]
    from collections import defaultdict
    dic=defaultdict(int)
    for word in corpus:
        if word in stop:
            dic[word]+=1
            
    top=sorted(dic.items(), key=lambda x:x[1],reverse=True)[:10] 
    x,y=zip(*top)

    plt.bar(x,y)
```

Your task is to plot the stopwords present in the news headlines.

```python
plot_top_stopwords_barchart(news['headline_text'])
```

It's time to visualize n-grams now.
![Image](https://i1.wp.com/neptune.ai/wp-content/uploads/trigrams.png?fit=1024%2C663&ssl=1)
 The function used to view n-grams is given to you:

```python
from sklearn.feature_extraction.text import CountVectorizer #scikit learn library: https://scikit-learn.org/stable/
import seaborn as sns # seaborn is a visualization library; more here: https://seaborn.pydata.org/introduction.html#:~:text=Seaborn%20is%20a%20library%20for,explore%20and%20understand%20your%20data.
def get_top_ngram(corpus, n=None):
    vec = CountVectorizer(ngram_range=(n, n)).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) 
                  for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:10]
```

You can view bi-grams like this:

```python
bigrams=get_top_ngram(news['headline_text'],2)[:10] 
x,y=map(list,zip(*bigrams)) 
sns.barplot(x=y,y=x)
```

Now, your task is to plot the first 20 4-grams...

```python
#your code here
fourgrams=get_top_ngram(news['headline_text'],4)[:20] 
x,y=map(list,zip(*fourgrams)) 
sns.barplot(x=y,y=x)
```

Now, try plotting the first 3 10-grams...

```python
#your code here
tengrams=get_top_ngram(news['headline_text'],10)[:3] 
x,y=map(list,zip(*tengrams)) 
sns.barplot(x=y,y=x)
```

<b>Wordcloud for headlines</b>

In this section, you have to create a wordcloud for the headline_text. We have given you a function that generates a list of all the words present in headlines. Let's call this list: <b>wordlist</b>

```python
from nltk.stem import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer 
from nltk import word_tokenize 
def preprocess_news(df):
    corpus=[]
    stem=PorterStemmer()
    lem=WordNetLemmatizer()
    stop=set(stopwords.words('english'))
    for news in df['headline_text']:
        words=[w for w in word_tokenize(news) if (w not in stop)]
        
        words=[lem.lemmatize(w) for w in words if len(w)>2]
        
        corpus.append(words)
    return corpus

#here is the list wordlist
wordlist=preprocess_news(news)
```

Now, generate the wordcloud for the headlines.

```python
#your code here
wc = WordCloud(width=3000, height=2000, random_state=1, background_color='white', collocations= False, stopwords=STOPWORDS).generate(str(wordlist))
plt.figure(figsize=(40,30))
plt.imshow(wc)
```
